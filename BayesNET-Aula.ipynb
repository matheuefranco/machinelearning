```
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
import pandas as pd
url = "https://raw.githubusercontent.com/matheuefranco/machinelearning/master/heart.csv"
dados = pd.read_csv(url)
dados.head()
```

```
#dados.describe()
```

```
x = dados.drop('num',axis=1) #features
y = dados['num'] #label ou target ou classificador
#print(y)
```

```
from sklearn.model_selection import train_test_split, cross_val_score
x_treino,x_teste,y_treino,y_teste =  train_test_split(x,y,test_size=0.3)
```

```
#Classificação com arvore de decisão
clf_ad = tree.DecisionTreeClassifier()
clf_ad = clf_ad.fit(x_treino,y_treino)
# Classificação com Bayes
clf_nb = GaussianNB()
clf_nb = clf_nb.fit(x_treino,y_treino)
```

```
preditos_ad = clf_ad.predict(x_teste)
preditos_nb = clf_nb.predict(x_teste)
real = y_teste.values
print("Preditos:",preditos_ad[0:20])
print("Real:    ",real[0:20])
```

```
from sklearn.metrics import accuracy_score
print("Arvore de Decisao:", accuracy_score(y_teste,preditos_ad))
print("Bayes:",accuracy_score(y_teste,preditos_nb))
```
